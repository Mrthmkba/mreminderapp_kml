```python
# Step 2: Identify columns with outliers
def find_outliers(df):
    outlier_columns = []
    for col in df.columns:
        if pd.api.types.is_numeric_dtype(df[col]):
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            if df[(df[col] < lower_bound) | (df[col] > upper_bound)].shape[0] > 0:
                outlier_columns.append(col)
    return outlier_columns

outlier_columns = find_outliers(df_f2_cleaned)
print("Columns with outliers:", outlier_columns)
```

```python
# Step 3: Remove outliers for 'e_cgpa' column
def remove_outliers_for_column(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df
# Remove outliers specifically for 'e_cgpa'
df_f2_cleaned1 = remove_outliers_for_column(df_f2_cleaned, 'e_cgpa')
```

Next we will do Data integration Part where we merge the smalldataset.xlsx containing Companies names and their ratings with our current dataset which we save as bigdata.xlsx

```python
import pandas as pd

# Load data from Excel files
cleandata_file = 'Cleandata.xlsx'
companiesdata_file = 'Companies.xlsx'

cleandata_df = pd.read_excel(cleandata_file)
companiesdata_df = pd.read_excel(companiesdata_file)

for index, row in cleandata_df.iterrows():
    # Check if the name exists in companiesdata_df
    if row['e_17_nama_syarikat'] in companiesdata_df['e_17_nama_syarikat'].values:
        # Get the corresponding rating from companiesdata_df and assign it to cleandata_df
        corresponding_rating = companiesdata_df.loc[companiesdata_df['e_17_nama_syarikat'] == row['e_17_nama_syarikat'], 'ratings_syarikat'].values[0]
        cleandata_df.at[index, 'ratings_syarikat'] = corresponding_rating
```
